```python
#!/usr/bin/env python
import pandas as pd

df = pd.read_csv('/data/serverless-ac21/YmlStats.csv',index_col=False)
df
```

# Functions

```python
print('Count of projects with X functions')
print(df.groupby(['FN Count']).size())

print('\nProjects with public functions: {}'.format(df[df['Public FN Count'] > 0].shape[0]))
print('Projects with private functions: {}'.format(df[df['Private FN Count'] > 0].shape[0]))

print('\nNumber of public functions: {}'.format(df['Public FN Count'].sum()))
print('Number of private functions: {}'.format(df['Private FN Count'].sum()))
```

# Number of Resources

```python
print('Projects that use resources: {}'.format(df[df['Resource Count'] > 0].shape[0]))
print('Projects that use resources with public function: {}'.format(df[(df['Resource Count'] > 0) & (df['Public FN Count'] > 0)].shape[0]))
print('Projects that use Private Resources: {}'.format(df[df['Private Resources'] > 0].shape[0]))
print('Projects that use Public RO Resources: {}'.format(df[df['Public RO Resources'] > 0].shape[0]))
print('Projects that use Public WO Resources: {}'.format(df[df['Public WO Resources'] > 0].shape[0]))
print('Projects that use Public RW Resources: {}'.format(df[df['Public RW Resources'] > 0].shape[0]))

print('\nNumber of private resources: {}'.format(df['Private Resources'].sum()))
print('Number of public RO resources: {}'.format(df['Public RO Resources'].sum()))
print('Number of public WO resources: {}'.format(df['Public WO Resources'].sum()))
print('Number of public RW resources: {}'.format(df['Public RW Resources'].sum()))

print('\nCount of resources by type')
print('Total Resources used: {}'.format(df['Resource Count'].sum()))
print('dynamodb Resources: {}'.format(df['dynamodb Resources'].sum()))
print('s3 Resources: {}'.format(df['s3 Resources'].sum()))
print('sns Resources: {}'.format(df['sns Resources'].sum()))
print('ssm Resources: {}'.format(df['ssm Resources'].sum()))
print('sqs Resources: {}'.format(df['sqs Resources'].sum()))
print('sdb Resources: {}'.format(df['sdb Resources'].sum()))
print('kms Resources: {}'.format(df['kms Resources'].sum()))
print('kinesis Resources: {}'.format(df['kinesis Resources'].sum()))
print('rds Resources: {}'.format(df['rds Resources'].sum()))
print('secretsmanager Resources: {}'.format(df['secretsmanager Resources'].sum()))
print('route53 Resources: {}'.format(df['route53 Resources'].sum()))
print('apigateway Resources: {}'.format(df['apigateway Resources'].sum()))
print('event Resources: {}'.format(df['event Resources'].sum()))

print("\nCount of projects with X resources")
df.groupby(['Resource Count']).size()
```
