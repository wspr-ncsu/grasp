```python
#!/usr/bin/env python
import pandas as pd

df = pd.read_csv('data/prolog-results.csv', index_col=False)
df
```

```python
df2 = pd.read_csv('data/parsed-permissions.csv', index_col=False,quotechar="'")
df = df.merge(df2, on='file')
df
```

# Functions

```python
print('Count of projects with X functions')
print(df.groupby(['total_fns']).size())


print('\nProjects with functions: {}'.format(df[df['total_fns'] > 0].shape[0]))
print('Projects with public functions: {}'.format(df[df['public_fns'] > 0].shape[0]))
print('Projects with private functions: {}'.format(df[df['private_fns'] > 0].shape[0]))

print('\nProjects with only public functions: {}'.format(df[(df['public_fns'] > 0) & (df['private_fns'] == 0)].shape[0]))
print('Projects with only private functions: {}'.format(df[(df['public_fns'] == 0) & (df['private_fns'] > 0)].shape[0]))

print('\nNumber of public functions: {}'.format(df['public_fns'].sum()))
print('\nNumber of public functions: {}'.format(df['public_fns'].sum()))
print('Number of private functions: {}'.format(df['private_fns'].sum()))
```

# Number of resources

```python
print('Projects that use resources: {}'.format(df[df['total_res'] > 0].shape[0]))
print('Projects that use resources with public function: {}'.format(df[(df['total_res'] > 0) & (df['public_fns'] > 0)].shape[0]))
print('Projects that use Private Resources: {}'.format(df[df['private_res'] > 0].shape[0]))
print('Projects that use Public RO Resources: {}'.format(df[df['public_r_res'] > 0].shape[0]))
print('Projects that use Public WO Resources: {}'.format(df[df['public_w_res'] > 0].shape[0]))
print('Projects that use Public RW Resources: {}'.format(df[df['public_rw_res'] > 0].shape[0]))

print('\nNumber of private resources: {}'.format(df['private_res'].sum()))
print('Number of public RO resources: {}'.format(df['public_r_res'].sum()))
print('Number of public WO resources: {}'.format(df['public_w_res'].sum()))
print('Number of public RW resources: {}\n'.format(df['public_rw_res'].sum()))

print('Projecs with public function and fully private resource: {}'.format(df[(df['public_fns'] > 0) &(df['private_res'] > 0)].shape[0]))
print('Projecs with public function and fully public resource: {}'.format(df[(df['public_fns'] > 0) &(df['public_rw_res'] > 0)].shape[0]))
print('Projecs with public function and public read resource: {}'.format(df[(df['public_fns'] > 0) &(df['public_r_res'] > 0)].shape[0]))
print('Projecs with public function and public write resource: {}'.format(df[(df['public_fns'] > 0) &(df['public_w_res'] > 0)].shape[0]))

print('\nCount of resources by type')
print('Total Resources used: {}'.format(df['total_resources'].sum()))
for res in [
    'kms', 'kinesis', 'secretsmanager', 'ses',
    'execute-api', 's3', 'ssm', 'es',
    'iam', 'cognito-idp', 'dynamodb', 'sqs',
    'states', 'logs', 'sns', 'rds', 'sdb'
    ]:
    print(f"{res} Resources: {df[res].sum()}")


print("\nCount of projects with X resources")
df.groupby(['total_res']).size()
```

# Policies

```python
print("Projects with '*' used for permissions: {}".format(df[df['full_wcp_any_r'] > 0].shape[0]))
print("Projects with '*' used for resources: {}".format(df[df['any_p_full_wcr'] > 0].shape[0]))
print("Projects with Partial Wildcard used for permissions: {}".format(df[df['half_wcp_any_r'] > 0].shape[0]))
print("Projects with Partial Wildcard used for resources: {}".format(df[df['any_p_half_wcr'] > 0].shape[0]))

print("Projects with '*' on '*': {}".format(df[df['full_wcp_full_wcr'] > 0].shape[0]))
print("Projects with 'x*' on 'x*': {}".format(df[df['half_wcp_half_wcr'] > 0].shape[0]))
print("Projects with '*' or 'x*' on '*' or 'x*': {}".format(df[df['any_wcp_any_wcr'] > 0].shape[0]))

print("\nProjects using only global policies: {}".format(
    df[
        (df['global_iam_stmts'] > 0) &
        (df['max_fn_iam_stmts'] == 0) &
        (df['fn_roles_used'] == 0)
    ].shape[0]))
print("Projects using only function policies: {}".format(
    df[
        (df['global_iam_stmts'] == 0) &
        ((df['max_fn_iam_stmts'] > 0) |
        (df['fn_roles_used'] > 0))
    ].shape[0]))
print("Projects using both global and function: {}".format(
    df[
        (df['global_iam_stmts'] > 0) &
        ((df['max_fn_iam_stmts'] > 0) |
        (df['fn_roles_used'] > 0))
    ].shape[0]))

print(f"Max number of stmts defined in project: {df['stmts_defined'].max()}")
print(f"Max number of stmts defined in project: {df[df['stmts_defined'] == 462]}")

```

# Events
